```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

import re
```

```python

url= "https://www.epicgardening.com/organic-heirloom-seed-companies-usa/"
response=requests.get(url)

doc = BeautifulSoup(response.text, "html.parser")
```

```python
container=doc.select_one("#post-1758 > div > div")
```

```python
# create an empty list
state_list=[]
# since I have to loop through twice, I will zip the two
for ele, idx in zip(container.find_all('h2'), container.find_all('p')[17:]):
#     create an empty dictionary
    state_dict ={}
    #grab state name
    state_dict["state_name"]=ele.text.strip()
    #grab state abbr
    state_dict["state_abbr"]=ele.find("a")['name']
    #overall description
    state_dict["descript"]=idx.text.strip()
    #grab link here. since some links are missing, let us use try and except
    try:
        state_dict["link"]=idx.find('a')['href']
    except:
        state_dict['link']= "No link"
        #grab company name, it is easier to do so here than extracting it from a dataframe
    try:
        state_dict['company_name']=idx.find('a').text.strip()
    except:
        state_dict['company_name']="no company"
#append list
    state_list.append(state_dict)
#put it in a dataframe
df=pd.DataFrame(state_list)
```

```python
#create a contact info column
df['contact_info']=df.descript.str.extract(r'(\d+-\d+-\d+)')
```

```python
#some cleaning 
df['cleaned_df']=df.descript.str.replace(r'(Contact: \d+-\d+-\d+ and Email:)', ""
                       ).str.replace("\xa0Contact:\xa0", ""
                                    ).str.replace(
    "\xa0Email", "").str.replace("email", "").str.replace("Email:", "").str.replace("protected", "")
    
```
